```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
sim_vtx = pd.read_csv("../cms_data/sim_vtx_z.csv")
rec_vtx = pd.read_csv("../cms_data/reco_vtx_z.csv")
track = pd.read_csv("../cms_data/tracks_z.csv")
```

```python
z = track.z[track.pt>0.5]

plt.plot(z,np.zeros(len(z)),"o",markersize=1,label="tracks_z")

sv_z = sim_vtx.z
plt.plot(sv_z,np.zeros(len(sv_z)),"x",markersize=10,color="green",label="sim_vtx_z")

v_z = rec_vtx.z
plt.plot(v_z,np.zeros(len(v_z)),"o",markersize=2,color="red",label="reco_vtx_z")

plt.xlim(-1,1)
plt.ylim(-0.001,0.001)
plt.legend()
```
